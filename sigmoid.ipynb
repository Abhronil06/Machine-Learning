```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
       return 1 / (1 + np.exp(-x))

# Generate a range of x values
# np.linspace creates evenly spaced numbers over a specified interval.
x = np.linspace(-10, 10, 500) # From -10 to 10 with 500 points

# Calculate the corresponding y values using the sigmoid function
y = sigmoid(x)

# Plot the sigmoid function
plt.figure(figsize=(8, 6)) # Optional: set the figure size
plt.plot(x, y, label='Sigmoid Function', color='blue')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.title('Plot of the Sigmoid Function')
plt.grid(True) # Adds a grid for better readability
plt.axhline(0.5, color='gray', linestyle='--', linewidth=0.8, label='y=0.5') # Add a line at y=0.5
plt.legend() # Show the legend with labels
plt.show()
```
