```python
def permute(s, answer=""):
    if len(s) == 0:
        print(answer)
        return
    for i in range(len(s)):
        ch = s[i]
        left_substr = s[:i]
        right_substr = s[i+1:]
        rest = left_substr + right_substr
        permute(rest, answer + ch)

permute("abc")
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.axes as ax
from IPython.display import clear_output
df = pd.read_csv("/home/lab4/Downloads/data_for_lr(2).csv")
df.head()
print(df.info())
df = df.dropna()
print(df.info())
duplicates_exist = df.duplicated().any()
print(duplicates_exist)
N = len(df)
x_train, y_train = np.array(df.x[0:500]).reshape(500, 1), np.array(df.y[0:500]).reshape(500, 1)

x_test, y_test = np.array(df.x[500:N]).reshape(N - 500, 1), np.array(df.y[500:N]).reshape(N - 500, 1)
plt.plot(x_train, y_train, "+", label="Training point")
plt.plot(x_test, y_test, "*", label="Testing point")
plt.xlabel("X-Test")
plt.ylabel("Y-Test")
plt.legend()
plt.show()
class LinearRegression:
    def __init__(self):
        self.Q0 = np.random.uniform(0, 1) * -1  
        self.Q1 = np.random.uniform(0, 1) * -1 
        self.losses = []  

    def forward_propogation(self, training_input):
        predicted_values = np.multiply(self.Q1, training_input) + self.Q0  
        return predicted_values

    def cost(self, predictions, training_output):
        return np.mean((predictions - training_output) ** 2)  

    def finding_derivatives(self, cost, predictions, training_input, training_output):
        diff = predictions - training_output
        dQ0 = np.mean(diff)  
        dQ1 = np.mean(np.multiply(diff, training_input))  
        return dQ0, dQ1

    def train(self, x_train, y_train, lr, itrs):
        for i in range(itrs):
            
            predicted_values = self.forward_propogation(x_train)

           
            loss = self.cost(predicted_values, y_train)
            self.losses.append(loss)

            
            dQ0, dQ1 = self.finding_derivatives(loss, predicted_values, x_train, y_train)

            
            self.Q0 = self.Q0 - lr * (dQ0)
            self.Q1 = self.Q1 - lr * (dQ1)

         
            line = self.Q0 + x_train * self.Q1
            clear_output(wait=True)
            plt.plot(x_train, y_train, "+", label="Actual values")
            plt.plot(x_train, line, label="Linear Equation")
            plt.xlabel("Train-X")
            plt.ylabel("Train-Y")
            plt.legend()
            plt.show()
        return (
            self.Q0,
            self.Q1,
            self.losses,
        )  



lr = 0.0001 
itrs = 30 
model = LinearRegression()
Q0, Q1, losses = model.train(x_train, y_train, lr, itrs)

for itr in range(len(losses)):
    print(f"Iteration = {itr+1}, Loss = {losses[itr]}")
     
y_pred = Q0 + x_test * Q1
print(f"Best-fit Line: (Y = {Q1}*X + {Q0})")

plt.plot(x_test, y_test, "+", label="Data Points")
plt.plot(x_test, y_pred, label="Predited Values")
plt.xlabel("X-Test")
plt.ylabel("Y-Test")
plt.legend()
plt.show()

def myfunc(x):
  return Q1 * x + Q0

predicted_value = myfunc(30)

print(predicted_value)
     
     
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.axes as ax
from IPython.display import clear_output
df = pd.read_csv("/home/lab4/Downloads/data_for_lr(2).csv")
df.head()
print(df.info())
df = df.dropna()
print(df.info())
duplicates_exist = df.duplicated().any()
print(duplicates_exist)
N = len(df)
x_train, y_train = np.array(df.x[0:500]).reshape(500, 1), np.array(df.y[0:500]).reshape(500, 1)

x_test, y_test = np.array(df.x[500:N]).reshape(N - 500, 1), np.array(df.y[500:N]).reshape(N - 500, 1)

print(x_train.shape)
print(x_test.shape)
plt.plot(x_train, y_train, "+", label="Training point")
plt.plot(x_test, y_test, "*", label="Testing point")
plt.xlabel("X-Test")
plt.ylabel("Y-Test")
plt.legend()
plt.show()

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
print((x_train.min(),x_train.max()))
plt.scatter(x_train, y_train)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Train Data')
plt.grid(True)  # Enable grid
plt.show()
model = LinearRegression()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
mse = mean_squared_error(y_test, predictions)
print(f'Mean squared error is: {mse:.4f}')
plt.plot(x_test, predictions, color='red', linewidth=2, label='Regression Line')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Model')
plt.legend()
plt.grid(True)
plt.show()
temp = np.array([[0.7,]])
print(model.predict(temp))
```
