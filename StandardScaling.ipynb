```python
data = [2, 4, 4, 4, 5, 5, 7, 9]

# Calculate mean and standard deviation
mean = sum(data) / len(data)

print("Mean:", mean)
```

Standard deviation - Wikipedia.jpg

```python

differences = [(x - mean) for x in data]

print("Differences:", differences)
```

```python
varience = sum([diff ** 2 for diff in differences]) / len(data)

print("Varience:", varience)
```

```python
std_dev = (varience) ** 0.5

print("Standard Deviation:", std_dev)
```

Statistics - Standard Normal Distribution.jpg

```python
# Calculate Z-scores
z_scores = [(x - mean) / std_dev for x in data]

# Print Z-scores
print("Z-Scores:", z_scores)
```

```python
mean = sum(z_scores) / len(data)
print(mean)



differences = [(x - mean) for x in z_scores]
std_dev = (sum([diff ** 2 for diff in differences]) / len(data)) ** 0.5

print(std_dev)
```
